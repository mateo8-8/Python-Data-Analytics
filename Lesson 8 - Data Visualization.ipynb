Scatter Plot (using Seaborn):

    Loaded the tips dataset (a built-in dataset in Seaborn).

    Created a scatter plot with:

        total_bill on the x-axis

        tip on the y-axis

        Points colored by sex

    Added a title "Tips Dataset".

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame({
    "Category": ["A", "B", "C"],
    "Value": [10, 20, 15]
})
```

Bar Chart

```python
plt.bar(df["Category"], df["Value"])
plt.title("Bar Chart")
plt.show()
```

Seaborn Scatter Plot

```python
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Tips Dataset")
plt.show()
```
